# Building and Training the Neural Network

This neural network is built with the TensorFlow Keras API

```python
import tensorflow as tf

from sklearn.model_selection import train_test_split
import numpy as np
import matplotlib.pyplot as plt

print(tf.__version__)
```

## Loading and Pre-processing Data

```python
labelArr = np.append(np.load("Data/Rotated/d20Rot1Through14Labels.npy"),
                     np.load("Data/Rotated/d20Rot15Through20Labels.npy"))
imgArr = np.append(np.load("Data/Rotated/d20Rot1Through14Images.npy"),
                   np.load("Data/Rotated/d20Rot15Through20Images.npy"), axis = 0)
imgArr.shape
```

```python
plt.figure()
plt.imshow(imgArr[500])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
#Randomly splitting the dataset into a training and testing set: 
train_images, test_images, train_labels, test_labels = train_test_split(imgArr, labelArr, test_size=0.20, random_state=42)
```

```python
print(train_images.shape)
print(len(train_labels))
print(test_images.shape)
print(len(test_labels))
```

```python
#Normalizing the training image pixel values from 0-1
train_images = train_images / 255.0
test_images = test_images / 255.0

#Labels are [1:20], shift left so that they are [0:19] to make the rest of the code easier
train_labels = train_labels - 1
test_labels = test_labels - 1
```

#### Verifying that pre-processing worked correctly

```python
plt.figure(figsize = (10,10))
for i in range(25):
    plt.subplot(5,5, i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap = plt.cm.binary)
    plt.xlabel(train_labels[i] + 1)
plt.show()
```

## Building the Model

```python
ConvModel = tf.keras.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation = 'relu', input_shape = (20,20, 1)),
    tf.keras.layers.MaxPooling2D((2,2)),
    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation = 'relu'),
    tf.keras.layers.Dense(20)
    
])

ConvModel.compile(optimizer = 'adam',
                 loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                 metrics = ['accuracy'])
```

## Training the Model

```python
conv_train_images = train_images.reshape(train_images.shape[0], 20, 20, 1)
ConvModel.fit(conv_train_images, train_labels, epochs = 5)
```

### Evaluate Accuracy

```python
conv_test_images = test_images.reshape(test_images.shape[0], 20, 20, 1)
conv_test_loss, conv_test_acc = ConvModel.evaluate(conv_test_images, test_labels, verbose = 2)
print('\nTest accuracy:', conv_test_acc)
```

### Make predictions

```python
#Convolution
conv_prob_model = tf.keras.Sequential([ConvModel, tf.keras.layers.Softmax()])
conv_predictions = conv_prob_model.predict(conv_test_images)
conv_predictions[1]
```

```python
print("CNN")
print("prediction: " + str(np.argmax(conv_predictions[0]) + 1))
print("actual: " + str(test_labels[0] + 1))
```

```python
def plot_image(i, predictions_array, true_label, img):
    true_label, img = true_label[i] + 1, img[i]
    plt.grid(False)
    plt.xticks([])
    plt.yticks([])

    plt.imshow(img, cmap=plt.cm.binary)

    predicted_label = np.argmax(predictions_array) + 1
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'

    plt.xlabel("{} {:2.0f}% ({})".format(predicted_label,
                                100*np.max(predictions_array),
                                true_label),
                                color=color)

def plot_value_array(i, predictions_array, true_label):
    true_label = true_label[i]
    plt.grid(False)
    plt.xticks(range(1, 21))
    plt.yticks([])
    thisplot = plt.bar(range(1, 21), predictions_array, color="#777777")
    plt.ylim([0, 1])
    predicted_label = np.argmax(predictions_array)

    thisplot[predicted_label].set_color('red')
    thisplot[true_label].set_color('blue')
```

```python
num_rows = 5
num_cols = 3
num_images = num_rows * num_cols
plt.figure(figsize = (2*4*num_cols, 2*num_rows))
for i in range(num_images):
    plt.subplot(num_rows, 2*num_cols, 2*i+1)
    plot_image(i, conv_predictions[i], test_labels, conv_test_images)
    plt.subplot(num_rows, 2*num_cols, 2*i+2)
    plot_value_array(i, conv_predictions[i], test_labels)
plt.tight_layout()
plt.show()
```
